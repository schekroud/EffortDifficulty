```python
alldata = np.zeros(shape = [subs.size, binmids.size, ntimes]) * np.nan
subcount = -1
for i in subs:
    subcount +=1
    print(f'working on ppt {subcount+1}/{subs.size}')
    
    #read in single subject data
    data = np.load(op.join(wd, 'data', 'tuningcurves', f's{i}_TuningCurve_mahaldists_binstep{binstep}_binwidth{binwidth}.npy'))
    bdata = pd.read_csv(op.join(wd, 'data', 'tuningcurves', f's{i}_TuningCurve_metadata.csv'))
    [ntrials, nbins, ntimes] = data.shape
    
    #get bin info
    _, binmids, binstarts, binends = TuningCurveFuncs.createFeatureBins(binstep, binwidth)
    binmidsrad = np.deg2rad(binmids) #get radian values for bin centres
    
    data = data * -1 #invert this, so more positive (larger) values = closer (mahalnobis distances are small when test is close to train)
    meandata = np.nanmean(data, axis=0)
    alldata[subcount] = meandata
```

```python
alldata.shape
```

```python
#for each subject, demean across bins per time point
gmean_tp = alldata.copy()
for isub in range(subs.size):
    tmpsub = alldata[isub].copy()
    for tp in range(ntimes):
        isnan = np.isnan(tmpsub[:,tp])
        imean = np.nanmean(tmpsub[:,tp])
        tmpsub[~isnan,tp] = np.subtract(tmpsub[~isnan, tp], imean) 
    gmean_tp[isub] = tmpsub.copy()
gmean = np.nanmean(gmean_tp, axis=0)
```

```python
fig = plt.figure(figsize = [6, 4])
ax = fig.add_subplot(111)
ax.imshow(gmean, aspect='auto', origin = 'lower', cmap = 'RdBu_r', interpolation='none',
         extent = [times.min(), times.max(), binmids.min(), binmids.max()])
ax.set_ylabel(f'orientation')
ax.set_xlabel('time relative to stim1 onset')
```

<font color='red'>

this is the tuning curve that you construct by taking the average 'tuning curve' across all trials of a subject and averaging to create a single subject tuning curve. This is then demeaned across bins per time point, within subject. this single-subject demeaned profile is averaged across subjects to construct the figure above

```python
tstarts = [-0.3, 0,   0.3, 0.6, 0.9, 1.2]
tends   = [ 0,   0.3, 0.6, 0.9, 1.2, 1.5]
plotcount = np.arange(len(tstarts))+1
fig = plt.figure(figsize = [15, 3])
for x in zip(plotcount, tstarts, tends):
    ax = fig.add_subplot(1,6,x[0])
    tinds = np.logical_and(np.greater_equal(times, x[1]), np.less_equal(times, x[2]))
    tmpdat = np.nanmean(gmean[:,tinds], axis=1) #average across time
    ax.plot(binmids, tmpdat)
    ax.set_xlabel('orientation')
    ax.set_title(f'distances {x[1]}$\\rightarrow${x[2]}s')
    ax.set_ylim([-0.15, 0.15])
fig.tight_layout()
```

<font color='red'>

if we take windows of time in this tuning curve, we can plot the distances across orientation bins to observe how the shape of this curve changes over time

from this, fairly easy to see that pre-stimulus there is no real shape to the distribution of distances across orientations. Over time, there develops some orientation preference (leading to the cosine shape), which starts to decline in shape after 1s from stimulus onset.
